## Data Observations

As the tumor volume increases so does the tumor weight for the capomulin regimen.

As time increases the tumor volume decreases under the capomulin regimen.

The median final tumor volume after being treated with the four regimens of interest is 45 mm^3.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_DF = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")
# Display the data table for preview
merge_DF
```

```python
# Checking the number of mice.
nummouse = len(merge_DF["Mouse ID"].unique())
print(f"There are a total of {nummouse} mice in the study.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse2 = merge_DF.loc[merge_DF.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
merge_DF[merge_DF["Mouse ID"]=="g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_DF = merge_DF[merge_DF["Mouse ID"].isin(mouse2)==False]
clean_DF
```

```python
# Checking the number of mice in the clean DataFrame.
num2mouse = len(clean_DF["Mouse ID"].unique())
print(f"There are {num2mouse} in the study.")
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
sumstatdf = clean_DF.loc[:, ["Mouse ID", "Drug Regimen", "Tumor Volume (mm3)"]]
sumstatdf
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean = sumstatdf.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
median = sumstatdf.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
variance = sumstatdf.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
standard_deviation = sumstatdf.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
SEM = sumstatdf.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# This method produces everything in a single groupby function 

stats = pd.DataFrame({
    "Mean TV":mean,
    "Median TV":median,
    "Variance": variance,
    "STDEV": standard_deviation,
    "SEM": SEM
})

stats_DF = stats.round(2)

stats_DF
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
#clean_DF.loc[:, "Mouse ID", "Timepoint"].groupby("Mouse ID").max()
maxtime = clean_DF.groupby("Mouse ID").max()["Timepoint"]

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint_time
max_time_study = pd.merge(maxtime, clean_DF, on=["Mouse ID", "Timepoint"], how="left")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimen= ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
filter_df = max_time_study.loc[:, "Drug Regimen"] == "Capomulin"
four_quantiles=[.25, .5, .75]
one_drug = max_time_study.loc[filter_df, "Tumor Volume (mm3)"]
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimen= ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for each_drug in regimen:
    #filter_df = max_time_study.loc[:, "Drug Regimen"]== "Capomulin"
    #filtering drugs
    filter_DF = max_time_study.loc[:, "Drug Regimen"]==each_drug
    four_quantiles=[.25, .5, .75]
    #new drug df
    one_drugTV = max_time_study.loc[filter_DF, "Tumor Volume (mm3)"]
    #append drug tv
    tumor_vol_list.append(max_time_study.loc[filter_DF, "Tumor Volume (mm3)"])
    #________
    quartiles = one_drugTV.quantile(four_quantiles)
    #df.loc(row,col)
    lowerq = quartiles.loc[.25]
    upperq = quartiles.loc[.75]
    iqr = upperq - lowerq
    lower_bound = lowerq-(1.5*iqr)
    upper_bound = upperq+(1.5*iqr)
        # Locate the rows which contain mice on each drug and get the tumor volumes
        # Determine outliers using upper and lower bounds
    drug_filter = (one_drugTV>upper_bound) | (one_drugTV<lower_bound)
    outlier_count = len(one_drugTV.loc[drug_filter])
    print(f"There are {outlier_count} outliers for {each_drug}.")
```

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
drug_filter = clean_DF.loc[:, "Drug Regimen"]=="Capomulin"
capomulin = clean_DF.loc[drug_filter,:]
drug_filter = capomulin.loc[:, "Mouse ID"]=="s710"
mouse_single = capomulin.loc[drug_filter, :]
mouse_single.plot("Timepoint", "Tumor Volume (mm3)")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment for Mouse S710")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
cap_avg = capomulin[["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]].groupby("Mouse ID").mean()
plt.scatter(cap_avg["Weight (g)"], cap_avg["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs. Mouse Weight")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(st.pearsonr(cap_avg["Weight (g)"], cap_avg["Tumor Volume (mm3)"])[0], 2)
print(f"Correltaion is {corr}")
#st.pearsonr(cap_avg["Weight (g)"], cap_avg["Tumor Volume (mm3)"])
```

```python
# y = M X + b
m, b, r, p, std = st.linregress(cap_avg["Weight (g)"], cap_avg["Tumor Volume (mm3)"])
#LinregressResult(slope=0.9544396890241045, intercept=21.552160532685015, 
#rvalue=0.8419363424694718, pvalue=1.3225722434712642e-07, stderr=0.1275435903320134)
y_pred = m*cap_avg["Weight (g)"]+b
plt.scatter(cap_avg["Weight (g)"], cap_avg["Tumor Volume (mm3)"])
plt.plot(cap_avg["Weight (g)"], y_pred)
plt.xlabel("Weight in Grams")
plt.ylabel("Tumor Volume Cubic Milimeters")
plt.title("Average Tumor Volume vs. Weight/Mouse")
plt.show()
```
